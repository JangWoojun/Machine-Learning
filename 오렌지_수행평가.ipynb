```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] =False
import seaborn as sns

orange = pd.read_csv("/content/drive/MyDrive/30619 장우준 인공지능/dataset/orange.csv")
orange
```

```python
# Color, Blemishes 열을 원 핫 인코딩 해줘야 됨
o_encoding = pd.get_dummies(
    orange, columns=["Color", "Blemishes (Y/N)"]
)
o_encoding
```

```python
X = o_encoding.drop('Variety', axis=1) # 독립변수
y = o_encoding['Variety'] # 종속변수
display(X, y)
```

```python
o_encoding1 = pd.get_dummies(orange, columns=["Color", "Blemishes (Y/N)", "Variety"])
```

```python
corr = X.corr()

plt.figure(figsize=(40,20))
sns.heatmap(corr, annot=True)
```

```python
columns = ["Size (cm)",
           "Weight (g)",
           "Brix (Sweetness)",
           "pH (Acidity)",
           "Softness (1-5)",
           "HarvestTime (days)",
           "Ripeness (1-5)",
           "Quality (1-5)"]
for i in columns:
    plt.figure(figsize=(10, 4))
    sns.histplot(o_encoding[i], kde=True)
    plt.title(f'{i} 분포')
    plt.xlabel(i)
    plt.ylabel('빈도수')
    plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# 3. 표준화 스케일링
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(X_train)
#독립변수만 스케일링(표준화, 정규화)
X_train_scaled = ss.transform(X_train)
X_test_scaled = ss.transform(X_test)

print(display(X_train_scaled))
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train_scaled, y_train)
```

```python
print(f"knn score(train) = {round(knn.score(X_train_scaled, y_train), 2)}%")
print(f"knn score(test) = {round(knn.score(X_test_scaled, y_test), 2)}%")
```

```python
k_range = range(1, 32, 2)

scores = []

for k in k_range:
  knn = KNeighborsClassifier(n_neighbors=k)
  knn.fit(X_train_scaled, y_train)
  scores.append(knn.score(X_test_scaled, y_test))

plt.plot(k_range, scores)
plt.xlabel("value of k")
plt.ylabel("score")
plt.grid()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train_scaled, y_train)
print(f"lr score(train) = {round(lr.score(X_train_scaled, y_train), 2)}%")
print(f"lr score(test) = {round(lr.score(X_test_scaled, y_test), 2)}%")
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

print(f"dt score(train) = {round(dt.score(X_train, y_train), 2)}%")
print(f"dt score(test) = {round(dt.score(X_test, y_test), 2)}%")
```

```python
# Color, Blemishes 열을 원 핫 인코딩 해줘야 됨
o_encoding = pd.get_dummies(orange, columns=["Color", "Blemishes (Y/N)", "Variety"])
o_encoding
```

```python
X = o_encoding.drop('Quality (1-5)', axis=1) # 독립변수
y = o_encoding['Quality (1-5)'] # 종속변수
display(X, y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# 3. 표준화 스케일링
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(X_train)
#독립변수만 스케일링(표준화, 정규화)
X_train_scaled = ss.transform(X_train)
X_test_scaled = ss.transform(X_test)

print(display(X_train_scaled))
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train_scaled, y_train)

print(f"knn score(train) = {round(knn.score(X_train_scaled, y_train), 2)}%")
print(f"knn score(test) = {round(knn.score(X_test_scaled, y_test), 2)}%")
```

```python
k_range = range(1, 32, 2)

scores = []

for k in k_range:
  knn = KNeighborsClassifier(n_neighbors=k)
  knn.fit(X_train_scaled, y_train)
  scores.append(knn.score(X_test_scaled, y_test))

plt.plot(k_range, scores)
plt.xlabel("value of k")
plt.ylabel("score")
plt.grid()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train_scaled, y_train)
print(f"lr score(train) = {round(lr.score(X_train_scaled, y_train), 2)}%")
print(f"lr score(test) = {round(lr.score(X_test_scaled, y_test), 2)}%")
```

```python
# Decision Tree, plot_tree 임포트, 모델 생성, 학습, 스코어 계산

from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

print(f"dt score(train) = {round(dt.score(X_train, y_train), 2)}%")
print(f"dt score(test) = {round(dt.score(X_test, y_test), 2)}%")
```

```python
dt1 = DecisionTreeClassifier(max_depth=3)
dt1.fit(X_train, y_train)

print(f"dt1 score(train) = {round(dt1.score(X_train, y_train), 2)}%")
print(f"dt1 score(test) = {round(dt1.score(X_test, y_test), 2)}%")
```

```python
# Decision Tree + Grid Search 사용
from sklearn.model_selection import GridSearchCV

params = {'max_depth' : range(1, 30)}
gs = GridSearchCV(DecisionTreeClassifier(), params)
gs.fit(X_train, y_train)
```

```python
#최적의 모델의 하이퍼 파라미터
print(gs.best_params_)
#최적의 모델도 저장
dt = gs.best_estimator_
print('train data:', dt.score(X_train, y_train))
print('test data:' ,dt.score(X_test, y_test))
```

```python
print(
    sum(gs.cv_results_['mean_test_score'])
    /
    len(gs.cv_results_['mean_test_score'])
)
```

```python
# 더 많은 하이퍼파라미터 조합
from sklearn.model_selection import GridSearchCV
params = {'max_depth' : range(3,10,1),
          'min_samples_split' : range(2,50,5),
          'min_samples_leaf' : range(2,50, 5)}
gs = GridSearchCV(DecisionTreeClassifier(), params)
gs.fit(X_train, y_train)
```
