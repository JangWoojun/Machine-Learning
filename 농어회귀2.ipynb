```python
import numpy as np
import pandas as pd
```

```python
perch = pd.read_csv("/content/drive/MyDrive/30619 장우준 인공지능/dataset/perch단순회귀.csv")
perch.sample(5)
```

```python
perch.describe()
```

```python
perch.info()
```

```python
X = perch.drop(columns=["Weight", "Species"])
y = perch["Weight"]
display(X, y)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_scaled = ss.fit_transform(X)
display(X_scaled)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
print(f"기울기: {lr.coef_}")
print(f"절편: {lr.intercept_}")
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
```

```python
lr_pred = lr.predict(X_test)

print(f"mae = {mean_absolute_error(y_test, lr_pred)}")
print(f"mse = {mean_squared_error(y_test, lr_pred)}")
print(f"rms = {np.sqrt(mean_squared_error(y_test, lr_pred))}")
print(f"mape = {mean_absolute_percentage_error(y_test, lr_pred)}")
print(f"r2 score= {r2_score(y_test, lr_pred)}")
print(f"lr test score = {lr.score(X_test, y_test)}")
```

```python
import matplotlib.pyplot as plt

c = range(0, 12)
plt.plot(c, y_test, c="r")
plt.plot(c, lr_pred, c="b")
plt.xlabel("index")
plt.ylabel("weight")
plt.show()
```

```python
residual = y_test - lr_pred
plt.plot(c, residual, c="g")
plt.xlabel("index")
plt.ylabel("residual")
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
dt = DecisionTreeRegressor()
dt.fit(X_train, y_train)
print(f"DTR train set: {dt.score(X_train, y_train)}")
print(f"DTR test set: {dt.score(X_test, y_test)}")
print(f"DTR depth: {dt.get_depth()}")
```

```python
dt_pred = dt.predict(X_test)

print(f"mae = {mean_absolute_error(y_test, dt_pred)}")
print(f"mse = {mean_squared_error(y_test, dt_pred)}")
print(f"rms = {np.sqrt(mean_squared_error(y_test, dt_pred))}")
print(f"mape = {mean_absolute_percentage_error(y_test, dt_pred)}")
print(f"r2 score= {r2_score(y_test, dt_pred)}")
print(f"dt test score = {dt.score(X_test, y_test)}")
```

```python
plt.figure(figsize=(40, 20))
plot_tree(dt, feature_names=X.columns)
plt.show()
```

```python
dt2 = DecisionTreeRegressor(max_depth=3)
dt2.fit(X_train, y_train)
print(f"DTR2 train set: {dt2.score(X_train, y_train)}")
print(f"DTR2 test set: {dt2.score(X_test, y_test)}")
```

```python
plt.figure(figsize=(40, 20))
plot_tree(dt2, feature_names=X.columns)
plt.show()
```

```python
c = range(0, 12)
pred = dt2.predict(X_test)
plt.plot(c, y_test, c="r")
plt.plot(c, pred, c="b")
plt.ylabel("Weight")
plt.show()
```

```python
dt3 = DecisionTreeRegressor(min_samples_split=6)
dt3.fit(X_train, y_train)
print(f"DTR3 train set: {dt3.score(X_train, y_train)}")
print(f"DTR3 test set: {dt3.score(X_test, y_test)}")
print(f"DTR3 depth: {dt3.get_depth()}")
```

```python
plt.figure(figsize=(40, 20))
plot_tree(dt3, feature_names=X.columns)
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    "max_depth": [3, 4, 5, 6],
    "min_samples_split": [6, 7, 8, 9],
    "min_samples_leaf": [5, 6, 7, 8],
}

grid_cv = GridSearchCV(DecisionTreeRegressor(), params)
grid_cv.fit(X_train, y_train)
print(f"최적의 하이퍼 파라미터: {grid_cv.best_params_}")
print(f"최적의 예측 정확도: {grid_cv.best_score_}")
```

```python
dt_final = grid_cv.best_estimator_
print(f"DTR Final train set: {dt_final.score(X_train, y_train)}")
print(f"DTR Final test set: {dt_final.score(X_test, y_test)}")
```

```python
dt_pred = dt_final.predict(X_test)

print(f"mae = {mean_absolute_error(y_test, dt_pred)}")
print(f"mse = {mean_squared_error(y_test, dt_pred)}")
print(f"rms = {np.sqrt(mean_squared_error(y_test, dt_pred))}")
print(f"mape = {mean_absolute_percentage_error(y_test, dt_pred)}")
print(f"r2 score= {r2_score(y_test, dt_pred)}")
```

```python
plt.figure(figsize=(40, 20))
plot_tree(dt_final, feature_names=X.columns)
plt.show()
```
