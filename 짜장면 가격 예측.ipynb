```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 3, 5])
y = np.array([88, 60, 55])
```

```python
plt.figure(figsize=(5, 5))
plt.plot(x, y, 'bo')
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x.reshape(-1, 1), y)
```

```python
plt.plot(x, y, "bo--")
plt.plot(x, model.predict(x.reshape(-1, 1)), "r")
```

```python
pre_x = int(input("스마트폰 사용시간: "))
y_hat = model.predict([[pre_x]])
print(pre_x, [pre_x], [[pre_x]])
print(f"{pre_x}시간 사용해서 {y_hat}점 예상해요")
```

```python
year = np.array(range(2000, 2020))
year
```

```python
price = np.array([58.530, 59.262, 62.758, 69.324, 70.953, 71.252, 72.190, 74.886, 84.681, 86.165, 87.240, 93.339,94.445,96.090,97.308,100.00,102.81,106.1,110.83,115.04])
price
```

```python
price = price / 100 * 4468
price
```

```python
x = year
y = price
model = LinearRegression()
model.fit(x.reshape(-1, 1), y)
```

```python
# 편향: bias
model.intercept_
```

```python
# 기울기: coefficient
model.coef_
```

```python
plt.figure(figsize=(7, 5))
plt.title("price index")
plt.plot(year, price, "bo--")
plt.plot(x, model.predict(x.reshape(-1, 1)))
plt.xticks(np.arange(1998, 2021, 2))
plt.show()
```

```python
y = model.predict([[2026]])
y
```

```python
x = int(input("궁금한 년도의 짜장면 가격"))
y = model.predict([[x]])
print(f"{x}년의 예상된 짜장면 가격은 {y}원")
```
