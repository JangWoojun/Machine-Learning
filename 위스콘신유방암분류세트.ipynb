```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
# 딕셔너리 형태
# sklearn에서 가져오는 모든 데이터셋은 딕셔너리 형태임

dataset = load_breast_cancer()
dataset.keys()
```

```python
print(dataset["DESCR"])
```

```python
# 딕셔너리 형태이므로 데이터프레임으로 변환
cancer = pd.DataFrame(data = dataset.data, columns=dataset.feature_names)
cancer.sample(5)
```

```python
# 타겟(label, class)추가
# 1: benign tumor: 약성 종양 - 비교적 성장 속도가 느리고 전이되지 않음
# 0: malignant tumor: 악성 종양 - 주위 조직에 침입하면서 빠르게 성장하고 확산, 생명 위협

cancer["target"] = dataset.target
cancer.head()
```
