```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 데이터 입력

grape = np.array([[19, 300, 0], [20,500,0], [18,400,0], [25,500,1],[27,700,1], [28, 600,1]])
```

```python
# 분류 알고리즘 2종

from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
# 데이터 전처리

X = grape[:, :2] # 전체 행 중 0~1열
y = grape[:, 2]

print(X, y)
```

```python
# knn 모델 생성

knn = KNeighborsClassifier()
```

```python
# knn 모델 학습

knn.fit(X, y)
```

```python
# knn 모델 정확도 확인

knn.score(X, y)
```

```python
# Logistic Regression 모델 생성, 학습, 정확도 확인을 해보세요

lr = LogisticRegression()
lr.fit(X, y)
lr.score(X, y)
```

```python
# coef: 기울기 / intercept: 절편

print(lr.coef_, lr.intercept_)
```

```python
# 예측도 해보자

print(lr.predict(X))
print(knn.predict(X))
```

```python
print(f"knn은 {'거봉' if knn.predict([[22, 600]]) == 1 else '포도'}")
print(f"lr는 {'거봉' if lr.predict([[22, 600]]) == 1 else '포도'}")
print(f"knn은 {'거봉' if knn.predict([[20, 500]]) == 1 else '포도'}")
print(f"lr는 {'거봉' if lr.predict([[20, 500]]) == 1 else '포도'}")
```
